```python
#basic Bokeh line graph

#import Bokeh
from bokeh.plotting import figure
from bokeh.io import output_file, show

#prepare some data
x=[3.0,7.5,9.0]
y=[3.0,6.0,10.0]

#prepare output file
output_file("Line.html")

#create a figure object
f=figure()

#create a line plot
#f.line(x,y)
f.triangle(x,y)
f.circle(x,y)

#write the plot in the figure object
show(f)

```

```python
#basic Bokeh line graph

#import Bokeh and pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

#prepare some data
df=pandas.read_excel("http://pythonhow.com/verlegenhuken.xlsx")
x=df['Temperature'].astype(float)/10
y=df['Pressure'].astype(float)/10

#create a figure object
p=figure(plot_width=500,plot_height=400,tools='pan',logo=None)
 
p.title.text="Temperature and Air Pressure"
p.title.text_color="Gray"
p.title.text_font="arial"
p.title.text_font_style="bold"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label="Temperature (Â°C)"
p.yaxis.axis_label="Pressure (hPa)"   

#create a line plot
p.circle(x,y,size=0.5)

#prepare output file
output_file("Pressureovertemp.html")

#write the plot in the figure object
show(p)

```

```python
#basic Bokeh line graph

#import Bokeh and pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

#prepare some data
df=pandas.read_csv("adbe.csv",parse_dates=["Date"])
#create a figure object
p=figure(x_axis_type="datetime",sizing_mode='scale_width')

#create a line plot
p.line(df["Date"],df["Close"])

#prepare output file
output_file("Timeseries.html")

#write the plot in the figure object
show(p)

```
